```python
import pickle as pkl

with open ("OneVsTwoEncoder/EvalLSTURIni.pkl","rb") as f:
    evalLSTURini = pkl.load(f)

with open ("OneVsTwoEncoder/EvalLSTURIni2.pkl","rb") as f:
    evalLSTURini2 = pkl.load(f)
```

```python
# Plot AUC, MRR, NDCG5

import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(1,3, figsize=(15,5))

# Display final results as a bar plot

ax[0].set_title('AUC')
ax[0].set_ylabel('AUC')
ax[0].set_xlabel('Epochs')
ax[0].plot(evalLSTURini['AUC'], label='One News Encoder')
ax[0].plot(evalLSTURini2['AUC'], label='Two News Encoders')

ax[1].set_title('MRR')
ax[1].set_ylabel('MRR')
ax[1].set_xlabel('Epochs')
ax[1].plot(evalLSTURini['MRR'], label='One News Encoder')
ax[1].plot(evalLSTURini2['MRR'], label='Two News Encoders')

ax[2].set_title('NDCG@5')
ax[2].set_ylabel('NDCG@5')
ax[2].set_xlabel('Epochs')
ax[2].plot(evalLSTURini['NDCG5'], label='One News Encoder')
ax[2].plot(evalLSTURini2['NDCG5'], label='Two News Encoders')

plt.legend()
plt.show()
```
