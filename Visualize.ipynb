```python
import pickle as pkl
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open("TrainingLoglr0001.pkl", 'rb') as f:
    lr0001= pkl.load(f)

with open("TrainingLoglr001.pkl", 'rb') as f:
    lr001 = pkl.load(f)
```

```python
fig, axs = plt.subplots(2,2 ,figsize=(15,8))

ax11 = axs[0,0]
ax12 = axs[0,1]
ax21 = axs[1,0]
ax22 = axs[1,1]

ax11.plot(np.linspace(0,5,len(lr001[2])),lr001[2], label = 'Training Loss lr = 0.001')
ax11.plot(lr001[3], label = 'Validation Loss lr = 0.001')

ax11.set_title('Loss lr = 0.001')
ax11.legend()

ax12.plot(np.linspace(0,5,len(lr0001[2])),lr0001[2], label = 'Training Loss lr = 0.0001')
ax12.plot(lr0001[3], label = 'Validation Loss lr = 0.0001')

ax12.set_title('Loss lr = 0.0001')
ax12.legend()

ax21.plot(lr001[0], label = 'Training AUC lr = 0.001')

ax21.set_title('AUC lr = 0.001')
ax21.legend()

ax22.plot(lr0001[0], label = 'Training AUC lr = 0.0001')

ax22.set_title('AUC lr = 0.0001')
ax22.legend()


plt.show()
```
