```python
import pickle as pkl
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open('MindTrain.pkl', 'rb') as f:
    Loss,AUC,MRR,_,_ = pkl.load(f)
```

```python
fig, axs = plt.subplots(1,1 ,figsize=(15,8))

axs.plot(np.linspace(0,10,len(Loss)),Loss, label = 'Training Loss')
#axs.plot(Loss_vali, label = 'Validation Loss')
axs.set_title('Loss')

axs.legend()

plt.show()
```

```python
fig, axs = plt.subplots(1,1 ,figsize=(10,6))

axs.plot(AUC, label = 'AUC')
axs.set_title('AUC')
axs.legend()

plt.show()
```

```python
fig, axs = plt.subplots(1,1 ,figsize=(15,8))

axs.plot(MRR, label = 'MRR')
axs.set_title('MRR')
axs.legend()

plt.show()
```
