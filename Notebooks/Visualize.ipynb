```python
import pickle as pkl
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open('TrainTransformer2.pkl', 'rb') as f:
    Loss_training,AUC,MRR,NDCG5,NDCG10,Loss_vali = pkl.load(f)

Loss_training_mean = np.mean(np.array(Loss_training).reshape(706,-1), axis=1)
```

```python
fig, axs = plt.subplots(2,1 ,figsize=(15,8))

axs[1].plot(np.linspace(0,20,len(Loss_training_mean)),Loss_training_mean, label = 'Training Loss',color = 'orange')
axs[0].plot(Loss_vali, label = 'Validation Loss')

axs[0].legend()
axs[1].legend()

plt.show()
```
