```python
import pickle as pkl
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open('EvaluationTranformerIniLSTUR.pkl', 'rb') as f:
    results = pkl.load(f)[0]

```

```python
Loss_training_mean = np.mean(np.array(results['Loss_training']).reshape(265,-1), axis=1)
Loss_vali = np.array(results['loss_vali'])
AUC = np.array(results['AUC'])
MRR = np.array(results['MRR'])
```

```python
fig, axs = plt.subplots(2,1 ,figsize=(15,8))

axs[1].plot(np.linspace(0,10,len(Loss_training_mean)),Loss_training_mean, label = 'Training Loss',color = 'orange')
axs[0].plot(Loss_vali, label = 'Validation Loss')

axs[0].legend()
axs[1].legend()

plt.show()
```

```python
fig, axs = plt.subplots(1,1 ,figsize=(10,6))

axs.plot(AUC, label = 'AUC')
axs.set_title('AUC')
axs.legend()

plt.show()
```

```python
fig, axs = plt.subplots(1,1 ,figsize=(15,8))

axs.plot(MRR, label = 'MRR')
axs.set_title('MRR')
axs.legend()

plt.show()
```
